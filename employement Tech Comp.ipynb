UNIVARIATE ANALYSIS

```python
# Female %
sns.histplot(df['Female %'], bins=10, kde=True)
plt.xlabel('Female %')
plt.ylabel('Companies')
plt.title('Distribution of Female Employees in Tech Companies')
plt.show()

```

Female %: "The histogram shows that most tech companies have around 25-30% female employees, indicating a concentration in this range"

```python
# Male %
sns.histplot(df['Male %'], bins=10, kde=True)
plt.xlabel('Male %')
plt.ylabel('Number of Companies')
plt.title('Distribution of Male Employees in Tech Companies')
plt.show()

```

Male %: "The histogram shows that most tech companies have around 70-75 % male employees, indicating a concentration in this range"

```python
# Black %
sns.histplot(df['% Black'], bins=10, kde=True)
plt.xlabel('Black %')
plt.ylabel('Number of Companies')
plt.title('Distribution of Black Employees in Tech Companies')
plt.show()



```

Black %: "The histogram shows that most tech companies have around 15-21% Black employees, indicating a concentration in this range"

```python
# Asian %
sns.histplot(df['% Asian'], bins=10, kde=True)
plt.xlabel('% Asian')
plt.ylabel('Companies')
plt.title('Distribution of Asian Employees in Tech Companies')
plt.show()

```

```python
# White %
sns.histplot(df['% White'], bins=10, kde=True)
plt.xlabel('% White')
plt.ylabel('Number of Companies')
plt.title('Distribution of White Employees in Tech Companies')
plt.show()
```

White %: "The histogram shows that most tech companies have around 50-60% white race employees, indicating a concentration in this range"

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x='Year', y='Female %', hue='Company', marker='o')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Female %')
plt.title('Percentage of Females Employed by Company Over Years')
plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False)
plt.tight_layout()

# Show the plot
plt.show()
```

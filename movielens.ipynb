```python
# Find the number of movies per genre in the movie dataset.
genre_columns = df.columns[3:]
genre_counts = df[genre_columns].sum()
print(genre_counts)
```

```python
#  How many movies have more than one genre such as Action, Drama,
genre_columns = df.columns[3:]

# Calculate the sum of genre columns for each movie
df['genre_count'] = df[genre_columns].sum(axis=1)

# Count the number of movies with more than one genre
movies_with_multiple_genres = df[df['genre_count'] > 1].shape[0]
print (movies_with_multiple_genres)
```

```python
# What are the top 25 best-rated movies with a rating count of more than 100?
movie_title = merge_df2['movie title']
rating_great = merge_df2['rating']
for m, r in zip(movie_title, rating_great):
    if r > 100:
        print(m)
    else: print('None')
```

```python
#  What is the relationship between user demographics and movie ratings?
sns.histplot(merge_df2['gender'], bins=10, kde=True)
plt.xlabel('gender')
plt.ylabel('rating')
plt.title('correlation between rating and gender')
plt.show()
```

```python
sns.histplot(merge_df2['occupation'], bins=10, kde=True)
plt.xlabel('occupation')
plt.ylabel('rating')
plt.title('correlation between rating and occupation')
plt.show()
```

```python
sns.histplot(merge_df2['age'], bins=10, kde=True)
plt.xlabel('age')
plt.ylabel('rating')
plt.title('correlation between rating and age')
plt.show()
```
